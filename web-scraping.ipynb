```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://webpages.charlotte.edu/mscipion/"

r= requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text, "lxml")
```

```python
table = soup.find("table", class_=table)
print(table)
```

```python
headers = table.find_all("th")
print(headers)
```

```python
titles = []

for i in headers:
    title = i.text
    titles.append(title)
    
print(titles)
```

```python
df= pd.DataFrame(columns=titles)
print(df)
```

```python
rows = table.find_all("tr")
print(rows) 
```

```python
for i in rows[1:]:
   print(i)   
```

```python
for i in rows[1:]:
   data = table.find_all("td")
   print(data)
```

```python
for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text for tr in data]     
    l =len(df)                         
    df.loc[l]= row                     

df.head(6)  
```
